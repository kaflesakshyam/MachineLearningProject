```python
#Arrange the complaint types according to their average
#calculating average request closing time by complaint types and borough
avg_closing_time = df.groupby(['Complaint Type', 'Location Type'])['Request_Closing_Time'].mean().reset_index()
avg_closing_time['Request_Closing_Hours'] = avg_closing_time['Request_Closing_Time'] / 3600  #converting time into hours

#pivot table
pivot_table = avg_closing_time.pivot(index='Complaint Type', columns='Location Type', values='Request_Closing_Hours')
pivot_table
```

```python
#plotting in graph
pivot_table.plot(kind='bar', figsize=(12, 7),width=1.85)
plt.title('Average Request Closing Time (hours) by Complaint Type and Location')
plt.ylabel('Average Closing Time (Hours)')
plt.xlabel('Complaint Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#testing whether the average respnse time across complaint types is similar or not
new_df = df.dropna(subset=['Complaint Type', 'Request_Closing_Time']) #ensuring missing values are handled

top_complaints = new_df ['Complaint Type'].value_counts().head(5).index 
ranfil = new_df[new_df['Complaint Type'].isin(top_complaints)]

ranfil.head()
```

```python
#preparing for anova testing
grouped = ranfil.groupby('Complaint Type')
groups = [group['Request_Closing_Time'].dropna() for _, group in grouped]
group_names = list(grouped.groups.keys())

#converting time into hours for all the groups
groups_in_hours = [group / 3600 for group in groups]
```

```python
import scipy.stats as stats

f_stat, p_value = stats.f_oneway(*groups_in_hours)
```

```python
# Display the results
print(f"\nANOVA Test Results:")
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python

ANOVA Test Results:
F-statistic: 1789.8124885501425
P-value: 0.0
```

```python
#testing whether the complaint type and location type are related
#Perform Chi-Square Test of Independence
from scipy.stats import chi2_contingency
top_complaints = new_df ['Complaint Type'].value_counts().head(10).index
latest_df = new_df [new_df ['Complaint Type'].isin(top_complaints)].dropna(subset=['Location Type'])

#creating a contingency table
contingency_table = pd.crosstab(latest_df['Complaint Type'], latest_df['Location Type'])
```

```python
#displaying the results
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
print(f"\nChi-Square Test Results:")
print("Chi-square Statistic:", chi2)
print("P-value:", p_value)
print("Degrees of Freedpm:", dof)
```
